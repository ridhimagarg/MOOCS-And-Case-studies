### Importing libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import keras
import theano
```

## Importing the dataset
### Checking of dataset and dependent variables(input)
### Identifying the input and output variables for the model

```python
## Importing the dataset
dataset = pd.read_csv('/Churn_Modelling.csv')
print(dataset)

## Generating the input variables and output for the model
X = dataset.iloc[:,3:13].values
y = dataset.iloc[:,13].values
```

```python
##Checking the input variables
print(X)
X.shape
```

## Preprocessing of dataset
### Label Encoder for country column and gender column of independent variable.
### Hot Encoder for handling the categorical data


```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
labelencoder_X_1 = LabelEncoder()
## Encoding of columns- "Country" and "Gender"
X[:,1] = labelencoder_X_1.fit_transform(X[:,1])
X[:,2] = labelencoder_X_1.fit_transform(X[:,2])

onehotencoder = OneHotEncoder(categorical_features = [1])
X = onehotencoder.fit_transform(X).toarray()
X = X[:,1:]
```

```python
##Checking the input variables
print(X.shape)
```

### Standardizing the data
Data will have zero mean and unit variance.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=0)

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

### Creating the classifier

```python
# Lets make Neural Network
from keras.models import Sequential
from keras.layers import Dense

#Creating the classifier

classifier = Sequential()
classifier.add(Dense(units = 6, kernel_initializer = 'uniform',activation='relu',
                     input_dim = 11))
classifier.add(Dense(units = 6, kernel_initializer = 'uniform',activation='relu'))
classifier.add(Dense(units = 1, kernel_initializer = 'uniform',activation='sigmoid'))
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy' ,metrics = ['accuracy'])

classifier.fit(X_train, y_train, batch_size = 10, nb_epoch =100)
```
